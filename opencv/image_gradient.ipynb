```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv.imread('/content/drive/MyDrive/datasets/opencv/messi5.jpg',0)
```

```python
laplacian=cv.Laplacian(img,cv.CV_64F)
laplacian_abs=np.absolute(laplacian)
laplacian_final=np.uint8(laplacian_abs)
```

```python
sobel_x=cv.Sobel(img,cv.CV_64F,1,0)
sobel_x_absolute=np.absolute(sobel_x)
sobel_x_final=np.uint8(sobel_x_absolute)
```

```python
sobel_y=cv.Sobel(img,cv.CV_64F,0,1)
sobel_y_absolute=np.absolute(sobel_y)
sobel_y_final=np.uint8(sobel_y_absolute)
```

```python
images=[img,laplacian_final,sobel_x_final,sobel_y_final]
titles=['Source','Laplacian','Sobel X','Sobel Y']
```

```python
for i in range(len(images)):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i],'gray')
  plt.title(titles[i])

plt.tight_layout()
plt.axis('off')
plt.show()
```
