```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv.imread('/content/drive/MyDrive/datasets/opencv/opencv-logo.png')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
kernel=np.ones((5,5),np.float32)/25
convolution=cv.filter2D(img,-1,kernel)
blur_kernel=(3,3)
blur=cv.blur(img,blur_kernel)
gaussian_blur=cv.GaussianBlur(img,blur_kernel,0)
median_blur=cv.medianBlur(img,3)
```

```python
images=[img,convolution,blur,gaussian_blur,median_blur]
titles=['Source','2D Convolution','Blur','Gaussian Blur','Median Blur']
```

```python
for i in range(len(images)):
  plt.subplot(3,2,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])

plt.tight_layout()
plt.show()
```
